### 결측치, 이상치 처리

#### 이상치 처리

```python
# 이상치를 결측치 처리하기! 
quartile_1 = data.quantile(0.25)
quartile_3 = data.quantile(0.75)
IQR = quartile_3 - quartile_1
tmp = data["SeriousDlqin2yrs"]
# print(tmp)
data_drop = np.where((data < (quartile_1 - 1.5 * IQR)) | (data > (quartile_3 + 1.5 * IQR)), None, data)
data_drop = pd.DataFrame(data_drop, columns=data.columns)
data_drop = data_drop.drop(columns=["SeriousDlqin2yrs"])
# print(tmp)
data_drop["SeriousDlqin2yrs"] = tmp
data_drop
```

```python
data_drop.isnull().sum()
```

#### 결측치 처리 

```python
# 1. 이상치를 처리하지 않은 경우

# print(data.isnull().sum(), data.shape)

data_n_median = data.fillna(data.median())
# print(data_n_median.isnull().sum(), data_n_median.shape)

data_n_mean = data.fillna(data.mean())
# print(data_n_mean.isnull().sum(), data_n_mean.shape)

data_n_drop = data.dropna()
# print(data_n_drop.isnull().sum(), data_n_drop.shape)
```

```python
# 2. 이상치를 null값으로 처리한 경우 


# print(data_drop.isnull().sum(), data.shape)

data_drop_median = data_drop.fillna(data_drop.median())
# print(data_drop_median.isnull().sum(), data_drop_median.shape)

data_drop_mean = data_drop.fillna(data_drop.mean())
# print(data_drop_mean.isnull().sum(), data_drop_mean.shape)

data_drop_drop = data_drop.dropna()
# print(data_drop_drop.isnull().sum(), data_drop_drop.shape)
```

### base model로 전처리 한 값 정하기

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

####  RandomForest

```python
def rf_test(data):
    y = data['SeriousDlqin2yrs']
    X = data.drop(columns='SeriousDlqin2yrs')
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
    
    rf = RandomForestClassifier(random_state=0)
    rf.fit(X_train, y_train)
    
    proba_train = rf.predict_proba(X_train)
    proba_test = rf.predict_proba(X_test)
    
    print("train : ", roc_auc_score(y_train, proba_train[:,1]))
    print("test : ", roc_auc_score(y_test, proba_test[:,1]) )
    print("-" * 100)
```

```python
rf_test(data_n_median)
rf_test(data_n_mean)
rf_test(data_n_drop)
rf_test(data_drop_median)
rf_test(data_drop_mean)
rf_test(data_drop_drop)
```

```python
# 데이터 선택 : data_n_mean (이상치 처리하지 않고, median으로 설정)
```

### 하이퍼파라미터 튜닝 

```python
from sklearn.model_selection import GridSearchCV

param = {
    'n_estimators' : [200, 300, 400, 500],
    'max_depth' : range(2, 7),
    'max_features' : range(4, 11)
}



rf = RandomForestClassifier(random_state=0)
gs = GridSearchCV(rf, param, scoring='roc_auc', cv=4, n_jobs=-1)

y = data_n_mean['SeriousDlqin2yrs']
X = data_n_mean.drop(columns='SeriousDlqin2yrs')
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
gs.fit(X_train, y_train)
```

```python
result_rf = pd.DataFrame(gs.cv_results_)
result_rf.sort_values('rank_test_score')
```

```python
best_model = gs.best_estimator_
proba_train_rf = best_model.predict_proba(X_train)[:, 1]
proba_test_rf = best_model.predict_proba(X_test)[:, 1]
```

```python
roc_auc_score(y_train, proba_train_rf), roc_auc_score(y_test, proba_test_rf)
```
